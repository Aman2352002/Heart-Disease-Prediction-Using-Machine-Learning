## Taking all features 

```python
x = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
x.head()
```

## Splitting the data into training and testing sets

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.20, random_state=42)
```

## Standardizing the features

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.transform(x_test)
```

## Training Gaussian Naive Bayes model

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(x_train_scaled,y_train)
```

```python
y_pred = model.predict(x_test_scaled)
```

## Accuracy

```python
from sklearn import metrics
metrics.accuracy_score(y_pred,y_test)
```

```python
x = df[['HighBP','Stroke','PhysActivity','HighChol','CholCheck','DiffWalk','Smoker','AnyHealthcare','NoDocbcCost','Sex','HvyAlcoholConsump']]
y = df['HeartDiseaseorAttack']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.20, random_state=42)
```

## Training Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
```

```python
logmodel.fit(x_train,y_train)
```

```python
y_pred = logmodel.predict(x_test)
```

## Accuracy

```python
from sklearn import metrics
metrics.accuracy_score(y_pred,y_test)
```

## Using  grid search CV for better hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'C': [0.1, 1, 10],  # Reduced values for C
    'solver': ['liblinear', 'saga'],  # Limited solvers  # Keeping only 'l2' penalty
    'max_iter': [100, 200],  # Reduced iterations
    'tol': [0.0001, 0.001]  # Reduced tolerance values
}
```

```python
search = GridSearchCV(estimator=logmodel, param_grid=param_grid, cv=5, scoring='accuracy')
```

```python
search.fit(x_train,y_train)
```

## Accuracy and Scores

```python
print(search.best_params_)
print(search.best_score_)
```
